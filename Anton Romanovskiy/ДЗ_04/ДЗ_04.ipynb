### Подсчет длин предложений и их колличества.
Занесение данных в обьект DataFrame, отброжение результатов в таблицу обьект-признак.
1. открытие файла для чтения 
2. запись текста в новую переменную
3. создание функции очищающей текст от знаков пунктуации 
4. отоброжение результатов 

```python
text = codecs.open("test.txt","r",encoding = "UTF8" ) #кодировка текста 
```

```python
string_text ="" 
for word in text:
    string_text += word.strip() # записываем текст в переменную, удаляя символы табуляции 
```

```python
def counter_words(string_text):  #функция удаляет все знаки пунктуации и переводы строки
    string_text = string_text.replace("\n", "")
    string_text = string_text.replace("\t","")
    string_text = string_text.replace(",", "")
    string_text = string_text.replace('"',"")
    string_text = string_text.replace("'","")
    string_text = string_text.replace(":", "")
    string_text = string_text.replace(";", "")
    string_text = string_text.replace("", "")
    string_text = string_text.replace("...", ".")
    sring_text = string_text.replace("!", ".") #меняем все знаки окончания предложения на точки 
    sring_text = string_text.replace("?", ".")
    string_text = string_text.lower() #меняет все заглавные буквы на маленькие 
    words = string_text.split('.') #создает список из слов по точкам
    return words 
```

```python
counter = pd.DataFrame(
            {'Предложение': counter_words(string_text) , 
             'длина предложенией': (list(map(len, counter_words(string_text))))})

counter
```

### Подсчет частоты встречаемости слов в загруженном файле:


```python
new_list_words = str(counter_words(string_text)) #преобразуем результат функции снова в строку
for _ in string.punctuation: #удаляем все знаки пунктуации после преобразования
    if _ in new_list_words:
        new_list_words = new_list_words.replace(_, '')

new_list_words = new_list_words.split(" ") #создаем список слов по пробелам 

for word in new_list_words:
    if len(word) <= 3:
        new_list_words.remove(word) #получаем список слов длинной больше трех#

count_word_dic = {}
for word in new_list_words:
    count_word_dic[word] = new_list_words.count(word)
```
