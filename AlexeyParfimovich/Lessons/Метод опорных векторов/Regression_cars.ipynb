## Модель цен на автомобили

Имеется набор данных с признаками автомобилей и ценами на них. Необходимо построить предсказание цены

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv( 'cars.csv', delimiter = '\t' )
data.head()
```

```python
data.shape
```

```python
data.info()
```

Самое простое предположение, чтобы было удобнее отобразить это посторить зависимости цены от пробега

```python
X = data[ ['Mileage'] ]
Y = data['Price']
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression(normalize=True)
```

```python
model.fit(X, Y)
```

```python
y_predict = model.predict(X)
```

```python
plt.plot(X, Y, 'o')
plt.plot(X, y_predict, 'k-')
plt.title("Снижение стоимости авто от пробега")
plt.xlabel("Пробег")
plt.ylabel("Стоимость")
plt.show()
```

```python
mean_absolute_error(Y, y_predict)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(Y, y_predict)
```

#### Строим линейную регрессию по всему датафрейму

```python
X = data.drop(['Price', 'Make', 'Trim', 'Type', 'Doors', 'Cylinder'], axis=1)
```

```python
X = X.astype({'Model': 'category'})
```

```python
X = pd.get_dummies(X)
```

```python
X.shape
```

```python
Y = data['Price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
X.head()
```

```python
model=LinearRegression(normalize=False)
```

```python
model.fit(X_train, y_train)
```

```python
y_predict_2 = model.predict(X_test)
```

```python
# Коэффициент детерминации
model.score(X_test, y_test)
```

```python
from sklearn.metrics import r2_score

print('MSE = ',mean_squared_error(y_test, y_predict_2))
print('Ср.абс.ош. = ',mean_absolute_error(y_test, y_predict_2))
print('r2 = ',r2_score(y_test, y_predict_2))
```

# LinearSVR

```python
from sklearn.svm import LinearSVR
from sklearn.model_selection import GridSearchCV
```

```python
grid = {
    'epsilon': [0],
    'C': [1, 3, 5]
}
gridsearch = GridSearchCV(LinearSVR(max_iter=1000, random_state=42), 
                          grid, cv=5, n_jobs=-1, verbose=1, refit=True)
gridsearch.fit(X_train, y_train)
```

```python
svr = LinearSVR()
svr.fit(X_train, y_train)
svr.score(X_test, y_test)
```

```python
gridsearch.best_params_
```

```python
gridsearch.score(X_test, y_test)
```

```python
model = LinearSVR(epsilon = 0, C = 5, max_iter=10000000, random_state=42).fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
y_predict_3 = model.predict(X_test)
```

```python
print('MSE = ',mean_squared_error(y_test, y_predict_3))
print('Ср.абс.ош. = ',mean_absolute_error(y_test, y_predict_3))
print('r2 = ',r2_score(y_test, y_predict_3))
```

```python
from sklearn.svm import SVR
```

```python
model = SVR(kernel = 'rbf', degree=2, epsilon = 0, C = 1).fit(X_train, y_train)
y_predict_4 = model.predict(X_test)
print('MSE = ',mean_squared_error(y_test, y_predict_4))
print('Ср.абс.ош. = ',mean_absolute_error(y_test, y_predict_4))
print('r2 = ',r2_score(y_test, y_predict_4))
```
