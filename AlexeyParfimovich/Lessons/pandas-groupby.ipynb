В целом, методы объекта pandas GroupBy делятся на несколько категорий:

- ***Методы агрегирования*** (также называемые методами сокращения ) объединяют множество точек данных в агрегированную статистику по этим точкам данных. Примером может служить сумма, среднее значение или медиана десяти чисел, где результатом является только одно число.

- **Методы фильтрации*** возвращаются к вам с подмножеством исходных файлов DataFrame. Чаще всего это означает использование .filter() для удаления целых групп на основе некоторой сравнительной статистики об этой группе и ее подтаблице. Также имеет смысл включить в это определение ряд методов, исключающих определенные строки из каждой группы.

- ***Методы преобразования** возвращают a DataFrameс той же формой и индексами, что и у оригинала, но с другими значениями. Как при использовании методов агрегации, так и при фильтрации результат DataFrameобычно будет меньше по размеру, чем вход DataFrame. Это не относится к преобразованию, которое преобразует сами отдельные значения, но сохраняет форму оригинала DataFrame.

- ***Мета-методы*** меньше связаны с исходным объектом, для которого вы вызвали .groupby(), и больше сосредоточены на предоставлении вам высокоуровневой информации, такой как количество групп и индексы этих групп.

- ***Методы построения графиков*** имитируют API построения графиков для pandas SeriesилиDataFrame , но обычно разбивают вывод на несколько подграфиков.

````python
title, ser = next(iter(df.groupby("outlet", sort=False)["title"]))
````

Дальше идет .str.contains("Fed"). Это возвращает логическое значение Series , Trueкогда заголовок статьи регистрирует совпадение при поиске. Конечно, первая строка начинается с "Fed official says weak data caused by weather,..." и загорается как True:

````python
ser.str.contains("Fed")
````

````python
 ser.str.contains("Fed").sum()
````

Результатом является количество упоминаний "Fed"Los Angeles Times в наборе данных. Та же процедура применяется для Reuters, NASDAQ, Businessweek и многих других.

````python
import timeit
import numpy as np
````

````python
def test_apply():
    """Version 1: using `.apply()`"""
    df.groupby("outlet", sort=False)["title"].apply(
        lambda ser: ser.str.contains("Fed").sum()
    ).nlargest(10)

def test_vectorization():
    """Version 2: using vectorization"""
    mentions_fed = df["title"].str.contains("Fed")
    mentions_fed.groupby(
        df["outlet"], sort=False
    ).sum().nlargest(10).astype(np.uintc)

print(f"Version 1: {timeit.timeit(test_apply, number=3)}")
print(f"Version 2: {timeit.timeit(test_vectorization, number=3)}")
````

При трехкратном запуске test_apply() функция занимает 8,54 секунды, а test_vectorization() всего 0,73 секунды. Это впечатляющая разница в процессорном времени для нескольких сотен тысяч строк. Подумайте, насколько существенной становится разница, когда ваш набор данных увеличивается до нескольких миллионов строк!
