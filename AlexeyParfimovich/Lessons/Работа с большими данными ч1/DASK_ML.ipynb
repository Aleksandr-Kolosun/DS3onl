```python
ddf = ddf.drop(['Unnamed: 0'], axis=1)
```

```python
ddf = ddf.categorize()
```

```python
ddf.head()
```

```python
ddf.info()
```

```python
prepared = dask.dataframe.reshape.get_dummies(ddf)
```

```python
prepared
```

```python
prepared.head()
```

```python
prepared.describe().compute()
```

```python
y = prepared.price.to_dask_array(lengths=True)
```

```python
arr = prepared.drop('price', axis=1).to_dask_array(lengths=True)
```

```python
from dask_ml.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(arr, y, test_size=0.1)
```

```python
from dask_ml.linear_model import LinearRegression,PoissonRegression
from dask_ml import xgboost

lr = LinearRegression(solver='lbfgs', max_iter=10)
lr_model = lr.fit(X_train, y_train)
```

```python
#p_model = PoissonRegression()
#p_model = lr.fit(X_train, y_train)
```

```python
y_predicted = lr_model.predict(X_test)
```

```python
from dask_ml.metrics import mean_squared_error, mean_absolute_error
from math import sqrt

sqrt(mean_squared_error(y_test, y_predicted))
mean_absolute_error(y_test, y_predicted)
```
